## One way frequency tables-pd.crosstab() 

```python
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt

os.chdir("C:/Users/sachi/Downloads")# main path of the floder
train=pd.read_csv("diamonds.csv")# filename
train.head()

```

```python
my_tab=pd.crosstab(index =train["clarity"],
                  columns="count")# classifaication the clarity as l1 and etc.. and giving its value
my_tab
```

```python
my_tab.plot.bar()
```

# Two way Frequency tables

```python
clarity_color_table=pd.crosstab(index =train["clarity"],
                  columns=train["color"])
clarity_color_table
```

# Chi-Square Test

```python
contingency_table=pd.crosstab(index =train["clarity"],
                  columns=train["color"],margins=True)
contingency_table
```

```python
from scipy.stats import chi2_contingency
chi2_contingency(pd.crosstab(train["clarity"],
                  train["color"]))[0:3]
```
